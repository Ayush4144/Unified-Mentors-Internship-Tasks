```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ElectricVehicle.csv")
```

```python
df
```

```python
print(df.info())
```

```python
df = df.dropna(how='all')
```

```python
df
```

```python
print("\nDescriptive statistics (numeric):\n", df.describe())
print("\nDescriptive statistics (categorical):\n", df.describe(include='object'))
```

```python
df['Year'] = df['Year'].astype('Int64')
```

```python
df
```

```python
print("\nchecking Missing values after cleaning:\n", df.isnull().sum())
```

```python
# 1. Total EV Sales over Years
sales_by_year = df.groupby("Year")["EV_Sales_Quantity"].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=sales_by_year.index, y=sales_by_year.values, marker="o")
plt.title("Total EV Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("EV Sales Quantity")
plt.show()
```

```python
# 2. Top 10 States by EV Sales
sales_by_state = df.groupby("State")["EV_Sales_Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_state.values, y=sales_by_state.index, palette="viridis")
plt.title("Top 10 States by EV Sales")
plt.xlabel("EV Sales Quantity")
plt.ylabel("State")
plt.show()
```

```python
# 3. Sales by Vehicle Category
sales_by_category = df.groupby("Vehicle_Category")["EV_Sales_Quantity"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_category.values, y=sales_by_category.index, palette="magma")
plt.title("EV Sales by Vehicle Category")
plt.xlabel("EV Sales Quantity")
plt.ylabel("Vehicle Category")
plt.show()
```

```python
# 4. Monthly Sales Trend (All Years Combined)
sales_by_month = df.groupby("Month_Name")["EV_Sales_Quantity"].sum()
month_order = ["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
sales_by_month = sales_by_month.reindex(month_order)
plt.figure(figsize=(10,5))
sns.lineplot(x=sales_by_month.index, y=sales_by_month.values, marker="o", color="teal")
plt.title("Monthly EV Sales Trend (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("EV Sales Quantity")
plt.show()
```

```python
# 5. EV Sales by Vehicle Type
sales_by_type = df.groupby("Vehicle_Type")["EV_Sales_Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_type.values, y=sales_by_type.index, palette="coolwarm")
plt.title("Top Vehicle Types by EV Sales")
plt.xlabel("EV Sales Quantity")
plt.ylabel("Vehicle Type")
plt.show()
```

```python
# 6. Heatmap: State vs Year (Top 10 States)
top_states = df.groupby("State")["EV_Sales_Quantity"].sum().sort_values(ascending=False).head(10).index
state_year_pivot = df[df["State"].isin(top_states)].pivot_table(
    index="State", columns="Year", values="EV_Sales_Quantity", aggfunc="sum", fill_value=0
)
plt.figure(figsize=(14,6))
sns.heatmap(state_year_pivot, cmap="YlGnBu", annot=False, cbar_kws={'label': 'EV Sales'})
plt.title("EV Sales by State and Year (Top 10 States)")
plt.xlabel("Year")
plt.ylabel("State")
plt.show()
```
