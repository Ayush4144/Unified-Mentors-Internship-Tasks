```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
df = pd.read_csv("netflix1.csv")
```

```python
df
```

```python
df = df.dropna(how='all')
```

```python
df
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
df['added_day'] = df['date_added'].dt.day
```

```python
df
```

```python
print(df.info())
```

```python
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = pd.to_numeric(df['duration_int'], errors='coerce')
```

```python
df
```

```python
# Fill missing duration values
median_movie_duration = df.loc[df['type']=="Movie", 'duration_int'].median()
df.loc[(df['type']=="Movie") & (df['duration_int'].isna()), 'duration_int'] = median_movie_duration
df.loc[(df['type']=="TV Show") & (df['duration_int'].isna()), 'duration_int'] = 1
```

```python
df
```

```python
df
```

```python
# 3. Content Release Trend Over the Years
yearly = df['release_year'].value_counts().sort_index()
yearly.plot(kind='line', marker='o')
plt.title("Content Release Trend Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# 4. Content Added Per Year (Based on date_added)
added_yearly = df['added_year'].value_counts().sort_index()
added_yearly.plot(kind='bar')
plt.title("Titles Added to Netflix Per Year")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# 5. Distribution of Movie Durations
sns.histplot(data=df[df['type']=="Movie"], x='duration_int', bins=30, kde=True, color="purple")
plt.title("Distribution of Movie Durations (minutes)")
plt.xlabel("Duration (minutes)")
plt.show()
```

```python
years = df['release_year'].to_numpy()
print("Earliest Release Year:", np.min(years))
print("Latest Release Year:", np.max(years))
```
